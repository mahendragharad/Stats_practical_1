```python
# creating the function with Q-Q plot to check whether the data is follows
# normal distribution or not 

import matplotlib.pyplot as plt
import scipy.stats as sts
import pylab
def plot_data(sample) :
    plt.figure(figsize = (10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample , kde = True )
    plt.subplot(1,2,2)
    sts.probplot(sample , dist = "norm" , plot = pylab)
    plt.show()
```

```python
plot_data(data3_copy)
# If the blue dots are completely
# comming on the straight line then we can say that
# the data is normally distributed 
```

```python
plot_data(s)
# If the blue dots are completely
# lies on the straight line then we can say that
# the data is normally distributed 
```

```python
plot_data(s2)
```

```python
plot_data(np.log(s2))
```
